```python
df = spark.read.json(
    "abfss://bronze@cmsstorageaccdiv.dfs.core.windows.net/cms/"
)
```

```python
df.printSchema()
```

```python
display(df.limit(10))
```

```python
from pyspark.sql.functions import col

cols_to_drop = [c for c in df.columns
                if c.endswith("_dual_color")
                or c.endswith("_description")
                or c.endswith("_change")]

df_silver = df.drop(*cols_to_drop)
```

```python
df_silver.printSchema()
```

```python
from pyspark.sql.functions import col

df_silver = (
    df_silver
    .withColumn(
        "number_of_users",
        col("number_of_users").cast("long")
    )
    .withColumn(
        "number_of_providers",
        col("number_of_providers").cast("long")
    )
    .withColumn(
        "number_of_dual_eligible_users",
        col("number_of_dual_eligible_users").cast("long")
    )
    .withColumn(
        "number_of_fee_for_service_beneficiaries",
        col("number_of_fee_for_service_beneficiaries").cast("long")
    )
    .withColumn(
        "total_payment",
        col("total_payment").cast("double")
    )
)
```

```python
df_silver.printSchema()
```

```python
df_silver.write \
    .mode("overwrite") \
    .parquet(
        "abfss://silver@cmsstorageaccdiv.dfs.core.windows.net/cms/"
    )
```

```python
silver_df = spark.read.parquet(
    "abfss://silver@cmsstorageaccdiv.dfs.core.windows.net/cms/"
)

display(silver_df.limit(10))
```
